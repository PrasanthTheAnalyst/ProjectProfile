```python
import pandas as pd
```

```python
df = pd.read_csv("/content/creditcard.csv")
```

```python
df
```

```python
df.drop('Time', axis = 1, inplace= True)
```

```python
df
```

```python
cases =len(df)
```

```python
cases
```

```python
non_fraud_cases = len(df[df['Class']==0])
```

```python
non_fraud_cases
```

```python
fraud_cases = len(df[df['Class']==1])
```

```python
fraud_cases
```

```python
fraud_percentage = round(fraud_cases/non_fraud_cases*100,2)
```

```python
fraud_percentage
```

Normalizing the values

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
amt = df['Amount'].values
```

```python
df['Amount'] = sc.fit_transform(amt.reshape(-1,1))
```

```python
df['Amount'].head
```

Data Training

```python
x = df.drop('Class', axis=1).values
y = df['Class'].values
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=0.2, random_state=0)
```

#Modeling

DecisionTree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree_model = DecisionTreeClassifier(max_depth= 4, random_state=1, criterion = 'entropy')
```

```python
tree_model.fit(x_train,y_train)
```

```python
tree_pred = tree_model.predict(x_test)
```

KNN Algorithm

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors = 5)
```

```python
knn.fit(x_train,y_train)
```

```python
knn_pred = knn.predict(x_test)
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr_pred =lr.predict(x_test)
```

SVM Algo

```python
from sklearn.svm import SVC
```

```python
svm = SVC()
```

```python
svm.fit(x_train,y_train)
```

```python
svm_pred = svm.predict(x_test)
```

Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(max_depth=4)
```

```python
rfc.fit(x_train,y_train)
```

```python
rfc_pred = rfc.predict(x_test)
```

Accuracy Score & Confusion Matrics


```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

Decision Tree Accuracy score

```python
accuracy_score(y_test,tree_pred)
```

Decision Tree Confusion Matrix

```python
confusion_matrix(y_test,tree_pred)
```

Knn algo accuracy score

```python
accuracy_score(y_test, knn_pred)
```

KNN Confusion Matrix

```python
confusion_matrix(y_test,knn_pred)
```

Logistic Regression Accuracy score

```python
accuracy_score(y_test,lr_pred)
```

Logistic regression Confusion Matrix

```python
confusion_matrix(y_test,lr_pred)
```

```python
accuracy_score(y_test,svm_pred)
```

Random forest Accuracy score

SVM Confusion Matrix

```python
confusion_matrix(y_test,svm_pred)
```

Random forest Accuracy score

```python
accuracy_score(y_test,rfc_pred)
```

```python
confusion_matrix(y_test,rfc_pred)
```
